```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
# Monthly data
months = ['January', 'February', 'March', 'April', 'May', 'June']
revenue = [5000, 5500, 6000, 6500, 7000, 7500]
expenses = [4000, 4500, 5000, 5500, 6000, 6500]

# Create a DataFrame from the data
data = {'Month': months, 'Revenue': revenue, 'Expenses': expenses}
df = pd.DataFrame(data)

# 1-Create the bar chart
plt.bar(df['Month'], df['Revenue'], label='Revenue', color='blue')
plt.bar(df['Month'], df['Expenses'], label='Expenses', color='red')
plt.xlabel('Month')
plt.ylabel('Amount ($)')
plt.title('Monthly Revenue vs Expenses')
plt.legend()
plt.show()

# 2-Create the line chart
plt.plot(df['Month'], df['Revenue'], marker='o')
plt.xlabel('Month')
plt.ylabel('Revenue ($)')
plt.title('Revenue Trend')
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
# Data for the pie chart
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 30, 20, 25]
explode_values = [0.1, 0.1, 0.1, 0.1]  # Equal explode values for each category

# Create the pie chart
plt.pie(values, labels=categories, explode=explode_values, autopct='%1.1f%%')
plt.title('Pie Chart with Equal Explode Values')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
np.random.seed(0)
x = np.random.rand(50)
y = np.random.rand(50)

# Create the scatter chart
plt.scatter(x, y, marker='o', alpha=0.4, s=50)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Chart')
plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(x, y, marker='o', alpha=0.4, s=50)
plt.xlabel('X', fontsize=16)
plt.ylabel('Y', fontsize=16)
plt.title('Scatter Chart', fontsize=16, color='red', loc='left')
plt.show()
```

## Q5: Create three line charts with Legends at left and colors.

```python
# Generate data for the line charts
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Create the line charts
plt.plot(x, y1, label='Sin', color='blue')
plt.plot(x, y2, label='Cos', color='green')
plt.plot(x, y3, label='Tan', color='red')

# Set the legend position to the left
plt.legend(loc='upper left')

# Set labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line Charts with Legends')

# Show the plot
plt.show()
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
# Generate data for the subplots
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Create subplots with 1 row and 3 columns
fig, axs = plt.subplots(1, 3, figsize=(20, 10))

# Plot the first chart
axs[0].plot(x, y1)
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')
axs[0].set_title('Chart 1')

# Plot the second chart
axs[1].plot(x, y2)
axs[1].set_xlabel('X')
axs[1].set_ylabel('Y')
axs[1].set_title('Chart 2')

# Plot the third chart
axs[2].plot(x, y3)
axs[2].set_xlabel('X')
axs[2].set_ylabel('Y')
axs[2].set_title('Chart 3')

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```
